```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Imports the HTML into python
page=requests.get('https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city')
soup = BeautifulSoup(page.text, 'lxml')
```

```python
df=pd.DataFrame({"Location":[''],"Society":[''],"type_of_apartment":[''], "Price":[''],"Area":[''],'Status':['']})
```

```python
posting=soup.find_all('div',class_="infoWrap")
for post in posting:
    try:
        price=post.find("td",class_="price").text
        area=post.find("td",class_='size').text
        status=post.find("tr",class_="hcol w44").text
        rooms=post.find("a",class_="typelink").text
        location=post.find("a",class_="loclink").text
        society=post.find("a",class_="projName").text
        df=df.append({"Location":location,"Society":society,"type_of_apartment":rooms,"Price":price,"Area":area,'Status':status},ignore_index=True)
    except:
        pass
```

```python
for i in range(1,200):
    next_page="https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?page="+str(i)
    page=requests.get(next_page)
    soup=BeautifulSoup(page.text,"lxml")
    posting=soup.find_all('div',class_="infoWrap")
    for post in posting:
        try:
            price=post.find("td",class_="price").text
            area=post.find("td",class_='size').text
            status=post.find("tr",class_="hcol w44").text
            rooms=post.find("a",class_="typelink").text
            location=post.find("a",class_="loclink").text
            society=post.find("a",class_="projName").text
            df=df.append({"Location":location,"Society":society,"type_of_apartment":rooms,"Price":price,"Area":area,'Status':status},ignore_index=True)
        except:
            pass
```
